```python
bins=[0,60,70,80,100]
labels=['low','Average', 'Good','Excellent']
degree_bin=pd.cut(df['degree_p'],bins=bins, labels=labels)
print(degree_bin)
df['degree_bin']=degree_bin
```

```python
sns.countplot(x='degree_bin', hue='status', data=df, palette='plasma')
plt.title('Placement By Acedemic Performance')
plt.xlabel('Number of Students')
plt.ylabel('Status')
plt.show()
```
